```python
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
# Read image from your local file system
# Can be others such as dogtest.jpg for example
original_image = cv.imread('stockimage2.jpg')
cv2_imshow(original_image)
```

```python
# Convert color image to grayscale for Viola-Jones
grayscale_image = cv.cvtColor(original_image, cv.COLOR_BGR2GRAY)

# Load the classifier and create a cascade object for face detection
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_alt.xml')

detected_faces = face_cascade.detectMultiScale(grayscale_image, 1.07)
# 1.3 originally

output_image = original_image
#make copy of original image to overlay with detections

# make blue rectangles over every detection
for (column, row, width, height) in detected_faces:
    cv.rectangle(
        output_image,
        (column, row),
        (column + width, row + height),
        (255, 0, 0),
        4
    )

#make smaller version
small = cv.resize(output_image, (0, 0), fx=0.3, fy=0.3)
```
